```python
import numpy as np
import os
import imageio
def generate_gif(env, Q, outputFile, max_steps = 100):
    frames = []
    obs, _ = env.reset(seed=100)  # fixed seed for consistency
    
    # Render a RGB frame
    frame = env.render()
    frames.append(frame)
    
    total_reward = 0
    
    for step in range(max_steps):
        # greedy action
        action = np.argmax(Q[obs])
        obs, reward, terminated, truncated, _ = env.step(action)
        total_reward += reward
        
        # Render RGB frame
        frame = env.render()
        frames.append(frame)

        if terminated or truncated:
            break

    print("total_reward obtained : ", total_reward)
    
    # ensure gifs folder exists
    os.makedirs("gifs", exist_ok=True)

    # save GIF
    gif_path = os.path.join("gifs", outputFile)
    imageio.mimsave(gif_path, frames, fps=4)
    print(f"Saved GIF: {gif_path}")
```

```python
# Question 2.4 :: Generate GIFS
env = DiagonalFrozenLake(render_mode = "rgb_array", start_state = (0, 5))

generate_gif(env, results["0:5:monte_carlo"]["Q"], "frozenlake_mc_(0,5).gif")
generate_gif(env, results["0:5:q_learning"]["Q"], "frozenlake_qlearning_(0,5).gif")

env = DiagonalFrozenLake(render_mode = "rgb_array", start_state = (0, 3))

generate_gif(env, results["0:3:monte_carlo"]["Q"], "frozenlake_mc_(0,3).gif")
generate_gif(env, results["0:3:q_learning"]["Q"], "frozenlake_qlearning_(0,3).gif")
```
